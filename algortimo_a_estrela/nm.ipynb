```python
import aestrela as a
import networkx as nx
import matplotlib.pyplot as plt
```

```python


# Criando um grafo
G = nx.Graph()

# Adicionando nós (opcional, porque ao adicionar arestas eles já são criados)
G.add_nodes_from(['A', 'B', 'C', 'D','E','F','G','H'])

# Adicionando arestas com pesos
# Sintaxe: (nó1, nó2, {'weight': valor})

# Primeira camada
G.add_edge('A', 'B', weight=8)
G.add_edge('A', 'G', weight=17)
# Segunda camada
G.add_edge('B', 'C', weight=12)
G.add_edge('B', 'D', weight=4)
G.add_edge('G', 'H', weight=9)
# Terceira camada
G.add_edge('D', 'E', weight=11)
G.add_edge('H', 'F', weight=20)
# Quarta camada
G.add_edge('E', 'F', weight=5)

# Posição dos nós para visualização
pos = nx.spring_layout(G)

# Desenhar nós e arestas
nx.draw(G, pos, with_labels=True, node_size=1200, node_color='lightblue', font_size=12, font_weight='bold')

# Extrair os pesos das arestas
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.Graph()

# Adicionando arestas com pesos
G.add_edge('A', 'B', weight=8)
G.add_edge('A', 'G', weight=17)
G.add_edge('B', 'C', weight=12)
G.add_edge('B', 'D', weight=4)
G.add_edge('G', 'H', weight=9)
G.add_edge('D', 'E', weight=11)
G.add_edge('H', 'F', weight=20)
G.add_edge('E', 'F', weight=5)

# Posicionamento manual para simular uma árvore
pos = {
    'A': (0, 4),
    'B': (-1, 3),
    'G': (1, 3),
    'C': (-1.5, 2),
    'D': (-0.5, 2),
    'H': (1, 2),
    'E': (-0.5, 1),
    'F': (0.5, 0)
}

# Desenhar nós e arestas
nx.draw(G, pos, with_labels=True, node_size=1200, node_color='lightblue', font_size=12, font_weight='bold')

# Extrair os pesos das arestas
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Certifique-se de ter instalado o pygraphviz ou pydot
# pip install pygraphviz
# ou
# pip install pydot

G = nx.Graph()
G.add_edge('A', 'B', weight=8)
G.add_edge('A', 'G', weight=17)
G.add_edge('B', 'C', weight=12)
G.add_edge('B', 'D', weight=4)
G.add_edge('G', 'H', weight=9)
G.add_edge('D', 'E', weight=11)
G.add_edge('H', 'F', weight=20)
G.add_edge('E', 'F', weight=5)

# Usando Graphviz para layout em árvore
pos = nx.nx_agraph.graphviz_layout(G, prog='dot')  # 'dot' é ótimo para hierarquias

nx.draw(G, pos, with_labels=True, node_size=1200, node_color='lightblue', font_size=12, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```
