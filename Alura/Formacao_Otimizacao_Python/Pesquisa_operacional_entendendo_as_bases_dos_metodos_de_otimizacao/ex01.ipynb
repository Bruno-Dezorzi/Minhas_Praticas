```python
origem = (0,0)
destino = (4,4)

enderecos = [(2,1),(2,2),(8,3),(5,3)]
```

```python
rota = [origem] + enderecos + [destino]
rota
```

```python
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean as distancia_eucliadiana
```

```python
tamanho_rota = len(rota)

distancia_percorrida = 0

for i in range(tamanho_rota):
    x, y = rota[i]
    if i == 0:
        cor = 'blue'
    elif i == tamanho_rota - 1: 
        cor = 'red'
    else: 
        cor = 'black'
    plt.scatter(x,y, color = cor)

    if i < tamanho_rota - 1:
        x1,y1 = rota[i + 1]
        dx = x1- x
        dy = y1 - y

        plt.arrow(x,y,dx,dy, color = 'black', head_width = 0.1)

        distancia_percorrida = distancia_percorrida + distancia_eucliadiana(rota[i],rota[i+1])

plt.title(f'Rota com distancia de {round(distancia_percorrida,2)}')
plt.show()
```

```python
def desenhar_rota(origem = None,destino = None,enderecos = None, rota = None):
    
    
    if rota == None:
         rota = [origem] + list(enderecos) + [destino]
    
    tamanho_rota = len(rota)

    distancia_percorrida = 0

    for i in range(tamanho_rota):
        x, y = rota[i]
        if i == 0:
            cor = 'blue'
        elif i == tamanho_rota - 1: 
            cor = 'red'
        else: 
            cor = 'black'
        plt.scatter(x,y, color = cor)

        if i < tamanho_rota - 1:
            x1,y1 = rota[i + 1]
            dx = x1- x
            dy = y1 - y

            plt.arrow(x,y,dx,dy, color = 'black', head_width = 0.1)

            distancia_percorrida = distancia_percorrida + distancia_eucliadiana(rota[i],rota[i+1])

    plt.title(f'Rota com distancia de {round(distancia_percorrida,2)}')
    plt.show()
```

```python
origem = (0,0)
destino = (4,4)

enderecos = [(8,3),(2,2),(2,1),(5,3)]
```

```python
desenhar_rota(origem= origem, destino= destino, enderecos= enderecos)
```

```python
def calcular_distancia_rota(rota):
    distancia = 0
    tamanho_rota = len(rota)
    
    for i in range(tamanho_rota) :
        if i < tamanho_rota - 1:
            distancia = distancia + distancia_eucliadiana(rota[i], rota[i+1])
    return distancia
```

```python
print(calcular_distancia_rota(rota))
```

```python
import itertools

menor_distancia = float('inf')
menor_rota = None

for permutacao in itertools.permutations(enderecos):
    rota = [origem] + list(permutacao) + [destino]
    distancia_rota = calcular_distancia_rota(rota)
    
    if distancia_rota < menor_distancia:
        menor_distancia = distancia_rota
        menor_rota = rota
        
print(menor_rota, menor_distancia)

desenhar_rota(rota = menor_rota)
```
