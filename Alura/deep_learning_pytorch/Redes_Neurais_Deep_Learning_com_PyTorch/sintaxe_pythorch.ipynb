Tipos de tensores

```python
import torch

lista = [[1,2,3],
         [4,5,6],
         [7,8,9]]

tns = torch.LongTensor(lista)

print(tns.dtype)
print(tns)
```

```python
import numpy as np
arr = np.random.randint(10,size=(3,4))
print(arr.dtype)

print(arr)
print(tns)

tns_numpy = torch.from_numpy(arr)

print(tns_numpy.dtype)
print(tns)
```

```python
tns1 = torch.ones(3,3)
tns2 = torch.zeros(3,3)
tns3 = torch.randn(3,3)

print(tns1)
print(tns2)
print(tns3) 
```

```python
arr = tns3.data.numpy()
print(arr.dtype)
arr
```

```python
print(tns3)
tns3[1,2] = 77
print(tns3[0:2])
```

Operações com vetores

```python
tns = torch.randint(low=0,high=100,size=(3,3)) 
tns1 = torch.randint(low=0,high=100,size=(3,3))

print(f"{tns}\n")
print(f"{tns1}\n")
print(f"Soma {tns + tns1}\n")
print(f"Sub {tns - tns1}\n")
print(f"Multi {tns * tns1}\n")
print(f"Div {tns / tns1}\n")
```

```python
print(tns1.size())
print(tns1.view(torch.float))
```

```python
tns1 = tns1.view(-1)

tns1
```
