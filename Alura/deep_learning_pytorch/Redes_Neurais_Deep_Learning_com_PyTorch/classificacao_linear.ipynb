```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -1
b = 4
c = 0.4

'''
ax + by + c = 0
ax + c = -by
ax + c / -b = -y
y = (-a*x -c) / b
'''
def plotline(a,b,c):
    x = np.linspace(-2,4,50)
    y = (-a*x -c) / b

    plt.ylabel('Y')
    plt.xlabel('X')
    plt.axvline(0,-1,1,color='k',linewidth = 1)
    plt.axhline(0,-2,4,color='k',linewidth = 1)
    plt.plot(x,y)
    plt.grid(True)
    plt.tight_layout()
    
plotline(a,b,c)
```

```python
# Ponto sobre a reta
p1 = (2,0.4)
ret1 =  a*p1[0] + b*p1[1] + c

# Ponto fora a baixo da reta 
p2 = (3,0.6) 
ret2 =  a*p2[0] + b*p2[1] + c

# Ponto fora a cima da reta
p3 = (2,0.8) 
ret3 =  a*p3[0] + b*p3[1] + c

plotline(a,b,c)
plt.plot(p1[0],p1[1],color ='b', marker='o')
plt.plot(p2[0],p2[1],color ='r', marker='o')
plt.plot(p3[0],p3[1],color ='g', marker='o')
```

```python
#ax + by + c = 0
print("%.2f" % ret1)

print("%.2f" % ret2)

print("%.2f" % ret3)
```

# Exerccio usando sklearn

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(46)

X ,Y = make_classification(n_features=2, n_redundant=0, n_informative=1,
                           n_clusters_per_class=1)
'''print(X[:5])
print(X.shape)
print(Y[:5])
print(Y.shape)'''

plt.scatter(X[:,0],X[:,1], marker='o',c=Y,
            edgecolors='k')
plt.grid()

p = X[10]
print(Y[10])
plt.plot(p[0],p[1], marker='^',markersize = 20)
plt.tight_layout()
```

```python
def plotmodel(w1,w2,b):
    
    plt.scatter(X[:,0],X[:,1], marker='o',c=Y,
            edgecolors='k')
    
    xmin,xmax = plt.gca().get_xlim()
    ymin,ymax = plt.gca().get_ylim()
    
    x = np.linspace(-2,4,50)
    y = (-w1*x -b) / w2

    plt.ylabel('Y')
    plt.xlabel('X')
    plt.axvline(0,-1,1,color='k',linewidth = 1)
    plt.axhline(0,-2,4,color='k',linewidth = 1)
    plt.plot(x,y)
    plt.grid(True)
    plt.tight_layout()
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    
w1 = 3.7
w2 = 1.35
b = -0.5
plotmodel(w1,w2,b)

p = (-1,1)
print((w1*p[0] + b) + w2*p[1])
```

```python
def classify(ponto,w1,w2,b):
    ret = w1 * ponto[0] + w2 * ponto[1] + b
    if ret >= 0:
        return 1,'yellow'
    else:
        return 0,'blue'
    
print(classify(p,w1,w2,b))
```

```python
p = (2,-1)
classe,cor = classify(p,w1,w2,b)
print(classe,cor)

plotmodel(w1,w2,b)
plt.plot(p[0],p[1], marker = '^',mfc = cor, markersize = 20)
```

```python
acertos = 0
amostras = len(X)

for k in range(amostras):
    categ, _ = classify(X[k],w1,w2,b)
    if categ == Y[k]:
        acertos += 1
      
acuracia = (acertos/amostras)*100        
print(f"Sua acurcia foi de {acuracia:.2f}%")
```
