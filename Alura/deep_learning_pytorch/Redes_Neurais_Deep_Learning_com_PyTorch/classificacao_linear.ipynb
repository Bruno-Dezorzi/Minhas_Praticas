```python
import numpy as np
import matplotlib.pyplot as plt

a = -1
b = 4
c = 0.4

# ax + by + c
# y = (-ax-c)/b

def plotline(a,b,c):
    x = np.linspace(-2,4,50)

    y = (-(a*x)-c)/b
    plt.axvline(0,-1,1,color='k',linewidth=1)
    plt.axhline(0,-2,4,color='k',linewidth=1)
    plt.plot(x,y)
    plt.grid()
    
plotline(a,b,c)
```

```python
plotline(a,b,c)
p1 = (2,0.4)
plt.plot(p1[0],p1[1], marker = 'o', color= 'b')
p2 = (1,0.6)
plt.plot(p2[0],p2[1], marker = 'o', color= 'r')
p3 = (3,-0.4)
plt.plot(p3[0],p3[1], marker = 'o', color= 'g')
```

```python
ret1 =  a*p1[0] + b*p1[1] + c
print("%.2f" % ret1)

ret2 =  a*p2[0] + b*p2[1] + c
print("%.2f" % ret2)

ret3 =  a*p3[0] + b*p3[1] + c
print("%.2f" % ret3)
```

```python
from sklearn.datasets import make_classification

np.random.seed(46)

X,Y = make_classification(n_features=2,n_redundant=0,n_informative=1,n_clusters_per_class=1)

print(X.shape,Y.shape)

plt.scatter(X[:,0],X[:,1], marker = 'o',c=Y,edgecolors='k')

p = X[10]
print(Y[10])
plt.plot(p[0],p[1],marker = '^', markersize = 20)
plt.tight_layout()
plt.show()
```

```python
def plotmodel(w1,w2,b):
    
    plt.scatter(X[:,0],X[:,1], marker = 'o',c=Y,edgecolors='k')
    
    xmin,xmax = plt.gca().get_xlim()
    ymin,ymax = plt.gca().get_ylim()
    
    x = np.linspace(-2,4,50)
    y = (-(w1*x)-b)/w2
    
    plt.axvline(0,-1,1,color='k',linewidth=1)
    plt.axhline(0,-2,4,color='k',linewidth=1)
    plt.plot(x,y)
    
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    plt.grid()
    plt.tight_layout()
    plt.show()
    
    
w1 = 5
w2 = 1
b = -0.4
plotmodel(w1,w2,b)


p = (-1,1)
print(w1 * p[0] + w2 * p[1] + b)
```

```python
def classify(ponto,w1,w2,b):
    ret = w1 * ponto[0] + w2 * ponto[1] + b
    
    if ret >= 0:
        return 1, 'yellow'
    else:
        return 0, 'blue'
    
p = (2,-1)

classe, cor = classify(p,w1,w2,b) 
print(classe,cor)

plt.figure(figsize=(10,8))
plt.plot(p[0],p[1],marker = '^',color = cor, markersize = 30)
plotmodel(w1,w2,b)
```

```python
plt.figure(figsize=(10,8))
azul = 0
amarelo = 0;



for i in range(Y.size):
    classe, cor = classify(X[i],w1,w2,b)
    plt.plot(X[i][0],X[i][1],marker = 'o', color= cor)
    
    if classe == 1:
        amarelo += 1
    else:
        azul += 1
        
xmin,xmax = plt.gca().get_xlim()
ymin,ymax = plt.gca().get_ylim()       
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)

x = np.linspace(-2,4,50)
y = (-(w1*x)-b)/w2
    
plt.axvline(0,-1,1,color='k',linewidth=1)
plt.axhline(0,-2,4,color='k',linewidth=1)
plt.plot(x,y)

plt.grid()
plt.tight_layout()
plt.show()
        
print(f"Amarelo = {amarelo} e Azul = {azul}")
```

```python
acertos = 0

for k in range(len(X)):
    categ, _ = classify(X[k],w1,w2,b)
    if categ == Y[k]:
        acertos += 1
        
print(f"{acertos} acertos de {len(X)} dados")
print(f"{acertos}% de acerto")
```
